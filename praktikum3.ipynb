```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sklearn
from sklearn.impute import SimpleImputer
warnings.simplefilter("ignore")
```

```python
df = pd.read_excel("datasetkesehatanjantung.xlsx")
df.describe()
```

```python
start_col = 2
end_col = 20

x = df.iloc[:, 0:4]
y = df.iloc[:, 4]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from imblearn.over_sampling import SMOTE
imputer = SimpleImputer(strategy='mean')
x = imputer.fit_transform(x)
smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
x_train = imputer.fit_transform(x_resampled)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_resampled)
```

```python
y_predict = model.predict(x_test)
akurasi = sklearn.metrics.accuracy_score(y_test, y_predict)
```

```python
print(x_test)
print(akurasi)
```

```python
result = pd.DataFrame({
    "Usia": x_test["Usia"],
    "BMI": x_test["BMI"],
    "Predicted": y_predict,
    "Actual": y_test
}) 
print(result)

result.to_excel("hasil_prediksi.xlsx", index=False)
```
